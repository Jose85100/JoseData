#Distribución binomial

```python
import matplotlib.pyplot as plt

# Valores posibles
valores = ['Cara', 'Sello']

# Probabilidad de cada valor
probabilidades = [0.5, 0.5]

# Crear el gráfico de barras de la función de probabilidad
plt.bar(valores, probabilidades)
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Función de Probabilidad - Moneda')
plt.show()
```

#Distribución hipergeométrica

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parámetros de la distribución hipergeométrica
N = 50  # Tamaño de la población
K = 20  # Número de elementos favorables en la población
n = 10  # Tamaño de la muestra

# Valores posibles
valores = np.arange(0, n+1)

# Calcula la función de probabilidad de la distribución hipergeométrica
probabilidades = hypergeom.pmf(valores, N, K, n)

# Crea el gráfico de barras de la función de probabilidad
plt.bar(valores, probabilidades)
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Función de Probabilidad - Distribución Hipergeométrica')
plt.show()
```
